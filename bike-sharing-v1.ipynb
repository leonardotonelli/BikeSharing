```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# EDA

```python
df = pd.read_csv("/kaggle/input/bike-sharing-demand/train.csv")
count = df["count"]
atemp = df["atemp"]
temp = df["temp"]
datetime = df["datetime"]
humidity = df["humidity"]
wind = df["windspeed"]
season = df["season"]
holiday = df["holiday"]
workingday = df["workingday"]
weather = df["weather"]
casual = df["casual"]
registered = df["registered"]
```

```python
datetime_mo = np.vectorize(lambda x: x[:7])(datetime)
df["temp_mo"] = datetime_mo
subsums_mo = df.groupby("temp_mo")["count"].sum()
plt.clf()
plt.figure(figsize=(18, 6))
subsums_mo.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Sum of Values')
plt.title('Sum of Values by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
day = np.arange(32)
temp_day = np.vectorize(lambda x: x[8:10])(datetime)
df["temp_dd"] = temp_day
plt.figure(figsize=(18, 6))
sns.boxplot(data = df, x="temp_dd", y="count", hue="season")
```
